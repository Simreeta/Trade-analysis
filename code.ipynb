```python
import pandas as pd
```

```python
#reads the xlsx file
df= pd.DataFrame(pd.read_excel("BankNiftyFutures_Data5Min.xlsx"))
df
```

# conversion to 15 mins time frame

```python
# converted to 15 mins timeframe
df['Open']=df['Open 5']#changed the name
df['High']=df['High 5'].rolling(4).max().shift(-3)#highest of 15 minutes
df['Low']=df['Low 5'].rolling(4).min().shift(-3)#lowest of 15 minutes
df['Close']=df['Close 5'].shift(-3)
df['Volume']=df['Volume 5']
df.drop(df.tail(3).index, inplace = True)
df=df.drop(['Open 5','Close 5','High 5','Low 5','Volume 5'],axis=1)
df
```

```python
df.drop(df.tail(3).index, inplace = True)
df
```

Vwap is calculated here

```python
def calculateVwap(data):
    data['TP'] = (data['High']+data['Low']+data['Close'])/3.0
    data['TradedValue']  = data['TP']*data['Volume']
    data['CumVolume'] = data['Volume'].cumsum()
    data['CumTradedValue'] = data['TradedValue'].cumsum()
    data['VWAP'] = data['CumTradedValue'] /data['CumVolume']
    return data

df = df.groupby('Final Date format').apply(calculateVwap)
df
```

Buy or sell according to Vwap is calculated here

# From here is way 1 of calculating profit

```python
def buy_sell(data1):
    if (data1['Close'] > data1['VWAP']):#if close is greater than vwap we buy
        return 'Buy'
    else:#if close is lesser than vwap we sell
        return 'Sell'
    return data1
        
df['Buy/Sell'] = df.apply(buy_sell,axis=1)
```

```python
def total(data2):
    if (data2['Close'] > data2['VWAP']):#if close is greater than vwap we buy
        return -1*data2['Close']# since we buy the money is spent so it is represented as negative for calculation of total
    else:#if close is lesser than vwap we sell
        return data2['Close']# since we buy the money is received so it is represented as positive for calculation of total
    return data2
        
df['Total'] = df.apply(total,axis=1)
df[['Date','Total','Buy/Sell']]
```

The above cell represents bought or sold

```python
#grouped by day
# profit or loss judged based on Buy or sell
#if buy 'close' amount is subtracted, if sell 'close' amount is added 
df1= df.groupby('Date')
df2=df1['Total'].sum()
df2
```

```python
# if the total of a year is positive, it is taken as profit and if it is negative it is taken as loss
# total of year
df3= df.groupby(df['Date'].dt.year)
df4=df3['Total'].sum().sort_values(ascending=False)#sorted in descending order
df_date=df3['Date']

# the values are ascending in descending order such that the year with maximum profit is the top most in the list
df4
```

Since df4 is sorted in descending order 2015 has the best performance

# Way 2 of calculating profit

```python
#if Close is greater than open, it is a profit else loss

df['Total_new']=df['Close']-df['Open']# here total is calculated based on opening and closing year
```

```python
def total_new(data3):
    if (data3['Total_new'] > 0):#if close is greater than vwap we buy
        return 'Profit'# since we buy the money is spent so it is represented as negative for calculation of total
    else:#if close is lesser than vwap we sell
        return 'Loss'# since we buy the money is received so it is represented as positive for calculation of total
    return data3
        
df['Profit/Loss'] = df.apply(total_new,axis=1)
df[['Date','Total_new','Profit/Loss']]
```

Date wise profit or Loss

```python
df5= df.groupby(df['Date'].dt.year)
df6=df5['Total_new'].sum().sort_values(ascending=False)#sorted in descending order

# the values are ascending in descending order such that the year with maximum profit is the top most in the list
df6
```

In this way 2017 has highest profit. Therefore, 2017 has best performance
